```python
df_train=pd.read_csv(ZipFile('/content/titanic.zip').open('train.csv'))
df_test=pd.read_csv(ZipFile('/content/titanic.zip').open('test.csv'))
```

```python
df_train['Age'].fillna(df_train['Age'].mean(), inplace=True)
df_train['Embarked'].fillna(df_train['Embarked'].mode()[0], inplace=True)
df_test['Age'].fillna(df_test['Age'].mean(), inplace=True)
df_test['Embarked'].fillna(df_test['Embarked'].mode()[0], inplace=True)
```

```python
g = {'male':0,'female':1}  # Gender
e = {'C':0,'Q':1,'S':2}    # Embarked from

df_train['Sex'] = df_train['Sex'].map(g)    ### Maps the number 0 with 'male' and the number 1 with 'female in the 'Sex' column of the dataframe
df_train['Embarked'] = df_train['Embarked'].map(e)

df_test['Sex'] = df_test['Sex'].map(g)    ### Maps the number 0 with 'male' and the number 1 with 'female in the 'Sex' column of the dataframe
df_test['Embarked'] = df_test['Embarked'].map(e)
```

```python
df_test
```

```python
X_train = df_train.select_dtypes(include='number')  ### Returns a dataframe containing only the numeric data-type columns
y_train = X_train['Survived']
X_train = X_train.drop(columns='Survived',axis=1)
```

```python
logistic.fit(X_train,y_train)
y_pred = logistic.predict(X_test)

print(logistic.score(X_train,y_train)*100)
```
